#### 1. Introduction 

> Over the last decade, there has been a concerted effort in public education to refocus on civics-related curricula. As of January 2022, 39 states have enacted formal civics education policies that include a stand-alone course, with 18 including some form of high-stakes testing (Education Commission of the States, 2016; Shapiro & Brown, 2018; Stern, 2021). Florida has been at the forefront of the movement since the passage of the Sandra Day O’Conner Act in 2010 (SDOA; The Sandra Day O’Conner Civics Education Act, 2010). The SDOA itself took several years and iterations before it was adopted as law (Education, 2008; Middle School Civics Education Assessment, 2009; The Sandra Day O’Conner Civics Education Act, 2010; Education [SPCC], 2007). Surely having a former U.S.  Supreme Court Justice as its namesake and supporter helped to finally push it through.

> Since the passage of SDOA, Florida has had three Republican governors, Charlie Crist (2006-2010), Rick Scott (2010-2018), and Ron DeSantis (2018-present). Having a conservative governor is more likely to lead to the positive adoption of increased civics education requirements (Fitzgerald et al., 2021). In fact, the current governor, Ron DeSantis has been a strong proponent of more rigorous civics policies for the state at both K-12 and higher education institutions (Florida Governor’s Press Office, 2021). 

> Perplexingly, given the favorable environment after 2010, only a minority of civics-related bills have passed both chambers and become laws. Why are some of these civics-related bills successful while others are dead on arrival? 

#### 2. Related Work 

> Ball (2008) describes how “governance is accomplished through the ‘informal authority’ of diverse and flexible networks, while government is carried out through hierarchies or specific bureaucracy” (p. 747). While this study acknowledges the presence of outside agents and their contribution to the process of creating, promoting, and influencing legislation, we choose to focus on the formal legislators and their characteristics. The extant political science literature supports the idea that the political capital of internal agents (i.e., state legislators) and the share of a political party’s control directly affect the potential success of proposed legislation (Berry & Berry, 1990; Garand, 1985; Park & Hassairi, 2021; Walker, 1969; Whitaker et al., 2012). Mohr’s (1969) work points to the interaction of motivations, obstacles, and resources in successful policy adoption. This is closely associated with the opening of policy windows that allow for certain types of policies to have more likelihood of making it to the agenda in specific political environments (John W. Kingdon, 1993; Kingdon, 1984). 




#### 3. Research Questions 

> The purpose of this study is to shed light on why certain proposed civics education legislation passes through the Florida legislature while others do not. This study is interested in the contents of the policy texts themselves as well as the characteristics of their sponsors. More specifically we hope to address the following four research questions:

1. What are the most common priorities, or topics, across Florida’s Civics Education bills?
2. Does alignment with priorities predict the legislative success of bills?
3. Does the primary sponsor predict the legislative success of bills? 
4. Does the primary sponsor’s status on an education committee or subcommittee predict the legislative success of bills?
 


#### 4. Methods 

Data
> To obtain a true count of all proposed bills that concern civics education, this study chooses to use the Florida House of Representatives bill viewer feature directly. The website allows the researcher to select both chambers, House and Senate, at the same time. It also provides all the proposed legislation for each session of the legislature based on the bill’s text. For this study, the terms “Civic”, “Civics”, “Education”, and “American” were used. Of the bills that surfaced under those key terms, the researcher searched the bill text itself for where the term was used. If the bill's language was concerning civics education, social studies education, and other related civics educational goals, the bill was included. This allowed for the inclusion of bills specifically mentioning civics education as well as those that affect civics curriculum and/or funding under the broader context of social studies or American history. Bills that mentioned civic organizations were excluded as were resolutions to honor people or places. This process was repeated for every session of the legislature from 2011 until 2022. 

> There were a total number of 89 bills that were included in the study for analysis (full sample). Of those 83 included added text (additon sample) to an existing statute while only 71 bills removed existing text (removed sample). With regard to chamber, 45 were Senate bills while 44 were from the House. The bill text from each of these bills was downloaded and converted into TXT files to perform the necessary analysis. The data included the bill’s: name, summary, passage, main and additional sponsor(s), the date they were filed, date of the last action, committee and subcommittee assignments, and the statutes affected if passed. To assess the potential impact of the sponsors, data was also collected about the sponsors: chamber, party, years in office, and their assignment on an education committee and/or subcommittee. All the data was collected from the Florida House website, which maintains a publicly accessible repository of all bills from each legislative session.

Text Preprocessing (Python)

> The three sets of texts were run through cleaning and pre-processing functions before being analyzed. This was done through Python and the codes are listed in Sections 5.1 and 5.2 of this study document. Section 5.1 merely brings in the needed packages. The cleaning in section 5.2 includes converting all text to lowercase, removing punctuation, replacing abbreviations, and removing any digits. Then all the stop words are removed from the corpus. Finally, all the text is tokenized, stemmed, and lemmatized so that it can be exported out for the analysis stage of teh study. 


Analysis Process (R)

>   We used the structural topic modeling approach (M. E. Roberts et al., 2016), which allows researchers to disambiguate the underlying common topics from the documents while simultaneously incorporating metadata into the analysis. First, we identified the optimal number of topics by comparing the statistical fit indices, such as likelihood, residuals, and semantic coherence (Figures 1,5,and 9 respectively). Next, we looked at the top 3 topics by proportion to see which topics were of greatest interest to policymakers regarding civics over the last 12 years (Figures 2,6,and 10 respectively). We then looked at the effects of the bill's topics on its passage rate (Figures 3,7,and 11 respectively). Lastly, a correlation was done to see if there were still strong connections between any groups of topics (Figures 4,8,and 12 respectively . When looking at the overall bill summaries, there was a good deal of connectivity, especially for Topic 23. For the added and removed text, the topics did not show this trend. This makes sense as the original bill summaries included all the text from all the bills and the high level of text overlapping across these 89 bills should have shown correlations.

##### 5.2. Preprocessing the Data in Python (Does not run on this form)

```r
#### Your code for the analysis will be provided here 

#### From WK4 NLP practice excersice 
class text_normalize():
    def import_data(self, column='summary'):
        import numpy as np
        import pandas as pd
        #self.data = []
        filename = 'FL_Civics_Bills_Simp.csv'
        data = pd.read_csv(filename)
        self.data = data[column]
        self.data = pd.DataFrame(self.data)
        self.data.columns =[column]
        self.data[column] = self.data[column].astype(str)
        return self.data

    def cleaning(self, string):
        import re
        string = string.lower() # step 1. lowercase

        punc = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~' # step 2. remove punctuations 
        string = string.strip(punc) 

        string = string.replace("can't", 'cannot') # step 3. replace abbreviated forms 
        string = string.replace("n't", ' not')
        string = string.replace("'ll", ' will')
        string = string.replace("'m", ' am')
        string = string.replace("he's", "he is")
        string = string.replace("it's", 'it is')
        string = string.replace("'re", " are")

        return re.sub(r"\d+", " ", string) # step 4. replace all the numbers to white space

    def stopwords(self) : 
        from nltk.corpus import stopwords
        stop_words = set(stopwords.words('english')) # import the stopwords 
        return list(stop_words)  

    
    def tokenize(self, string, stop=True):
        import nltk
        tokens = nltk.word_tokenize(string)
        if stop==True:
            stop_words = self.stopwords()
            return [i for i in tokens if i not in stop_words]
        else: 
            return tokens 

    def stemming(self, tokens):
        import nltk
        stemmer = nltk.stem.PorterStemmer()
        return [stemmer.stem(token) for token in tokens]
    
    def lemmatize(self, tokens):
        import nltk
        lemmatizer = nltk.stem.WordNetLemmatizer()
        return [lemmatizer.lemmatize(token) for token in tokens]
    
    def preprocess(self, column):
        self.df = self.import_data(column)
        self.df[column] = self.df[column].apply(self.cleaning) # cleaning applied to the summary column
        #self.df[column] = self.df[column].apply(self.stopwords) # stopwords applied to summary column
        
        
        self.df['tokens'] = self.df[column].apply(self.tokenize, stop=True) # tokenization applied to the cleaned summary 
        self.df['stem'] = self.df.tokens.apply(self.stemming)
        self.df['lemma'] = self.df.tokens.apply(self.lemmatize)

        return self.df 

normalizer = text_normalize() # call the text_normalize function and name it normalizer
df1 = normalizer.preprocess('summary')

# Save the cleaned and normalization data

df2 = normalizer.preprocess('addition')
df3 = normalizer.preprocess('removal')


# export the cleaned and normalization data for analysis
df1.to_csv("Clean_Sum.csv")
df2.to_csv("Clean_Add.csv")
df3.to_csv("Clean_Rem.csv")


####################################################
```

#### 6. Preliminary Results

#### 6.1. Number of Topics and Their Top Terms

#### RQ1: What are the most common priorities, or topics, across Florida’s Civics Education bills?

Bill Summary
> Based on the Held-Out Likelihood and Residuals seen in Figure 1, it was decided that 23 topics was the best fit. The full 23 topics are displayed by their proportions. As shown in Figure 2, the top three topics were Numbers 10, 16, and 13. Topic 10 includes stems like, "credit", "literaci", and "financ", which allow us to interpret this topic as part of bills that focused on fiancial literacy standards as a broader understanding of social studies. Topic 16 includes "civic", "engag", "practicum", and "nonpartisan", which can be thought to focus on the principles of active learning in civics courses. In this case, active learning refers to activities done to extend learning in the class such as protesting, calling a Congressperson, mock elections, or debates. Finally, topic 13 includes the use of "materi", "instruct", "meet", and "adopt", which focus on the organization and structure of civics education standards and their related curriculum. 

Bill Addition
> For the text that was added to a statute we also see a topic number of 23. Figure 5 shows the corresponding Held-Out  of 25 and Residuals of 22. Since these were not the same, a decision was made to chose a number inbetween them. As shown in Figure 6, the top three topics were Numbers 13, 22, and 9. Topic 13 includes terms such as "financ", "literaci", and "money" which aligns again with the focus on financial literacy. Topic 22 includes ""civic", "engag", "commun", and "activ" refering again to the active learning principles expected to be used in a civics course. Lastly, topic 9 includes "materi", "instruct", and "adopt" aligning itself to the organization and structure of the course.  

Bill Removal
> Figure 9 provides the Held-Out Likelihood and Residuals for the text that was removed from a statute. The Held-Out Likelihood reflects 17 topics and the Residuals suggest 15. For this study, the decision was made to create 16 topics. The top three topics for removal were Numbers 12, 10, and 5. Topic 12 once again aligns with the financial literacy push for social studies overall, containing terms such as "financ", "loan", and "econom". Topic 10 includes "assess", "grade", and "eoc" which is more aligned with changes in the assessments being used to evaluate civics instruction rather than the previous intructional materials focus. Topic 5 contains "skill", "standard", "workforc", adn "applic" which seem to suggest removing career preparation language from the social studies standards and expectations.





#### 6.2. Topics Effect on Passage Rate

#### RQ2: Does alignment with priorities predict the legislative success of bills?

Added Text

> Figure 7 showcases the effect of adding a top topic on the overall passage rate for a bill. The centers of two of the three topics do fall on the right side of the line, implying passage. We also see that for each of the three top topics, the confidence intervals touch "0.0", which means that they are not statistically significant. Unfortunately, they are not far enough into the passing side of the figure to claim that they are making a making any difference in the chances that a proposed bill will be passed into law.


Removed Text

> Figure 11 showcases the effect of removing a top topic on the overall passage rate for a bill. Again we see that two of the three top topics fall on the right side of the figure. Topic 5, however still has a confidence interval that touches "0.0" still making it statisitically insignificant. On the bright side, Topic 12 is far enough over to avoid touching "0.0" thus being the only node that is significant. This means that removing text found in topic 12 may actually lead to a proposed bill becoming a law!



#### 7. Discussion, Limitations, and Implications

Discussion
> The current study does not provide as clear of a results as one would have liked. While there are some terms that when being removed from a statute, made those bills successful, these terms are often found in the language of unsuccessful bills alike. In essence, the results are describing a situation in which the usage of certain terms, like those related to financial literacy, could both increase or decrease its passage likelihood. This is fundamentatlly problematic, even if we are seeing initial evidence to support it. Due to this conflict, it makes too noisy at this point to claim that subtracting text from removal topic 12 is genuinely increasing passage rates. Further research into research questions 3 and 4 will hopefully provide a more nuanced understanding of these inital results. Perhaps when accounting for the political party of a congressperson or their status on a committee, the observed significance will no longer hold.

Limitations
> One clear limitation of this study is that it lacks generalizability to other states as well as non-education topics. Since the study only focused on civics-related legislation in Florida and the sample size was under 100 proposed bills, we cannot claim that these results would hold for another state or subject matter. There is also a large amount of overlap due to the bicameral nature of the legislature. The study would be more generalizable if expanded by including all 50 states plus D.C. Without a direct repository of these proposed bills, however, this would be quite a large undertaking and become untenable.

> The Republican Party has maintained control over the Governorship as well as both chambers of the legislature throughout our study period. This does not allow us to analyze policy adoption variations under the leadership of each party.

> Lastly, although steps were taken to validate the sample data and the results from the topic modeling, there is always room for error with any analysis based on machine learning, text mining, and big data. There seems to be inconsistancy between the number of tokens found in the additional text when compared to the summary text. In principle, the bill summaries are longer and contain more words that any added or removed text. Somehow, the automated process used to bring in the inital text samples shows a higher number of tokens in the added text than the summery text. Further investigation is needed to understand this inconsistency as well as its potential bias to the current results. 

Implications
> Having knowledge and understanding concerning what language is most, or least, likely to lead to legislative success can be beneficial to Florida’s house and senate members, actively engaged citizens, policy entrepreneurs, and other policy hopefuls. It also provides an opportunity to discover changes in the contextual nature of the legislative body across sessions. We can determine if what legislators promote publicly during campaign seasons matches the priorities of that legislative body during their session.
